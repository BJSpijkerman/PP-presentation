```python
var('z L dL m n')

assume(L > 0)
assume(dL != 0)
assume(m != n*(L/(L + dL)))

k1 = m*pi/L
k2 = n*pi/(L + dL)

Delta_k = k1 - k2
Sigma_k = k1 + k2

I = 1/2 * (
    integral(cos(Delta_k*z), z, -L/2, L/2)
    + integral(cos(Sigma_k*z), z, -L/2, L/2)
)

I_2 = integral(cos(k1*z)*cos(k1*z), z, -L/2, L/2)
I_3 = integral(cos(k2*z)*cos(k2*z), z, -(L + dL)/2, (L + dL)/2)


show(simplify(I))
show(simplify(I_2))
show(simplify(I_3))
print(" ")
print("-------------------------")
print(" ")
T = simplify(I/sqrt(I_2*I_3))
show(T)
```

```python
T_mn = simplify(T.subs({m:1, n:2}))
show(T_mn)
print(latex(T_mn))
```

```python
# Use the same mode index to check for 100% overlap at dL=0
T_mn = T.subs({m:1, n:1}) 

# Correct normalization for Power Overlap (Square of the coefficient)
# Or use T_coeff = I / sqrt(I_2 * I_3) for the amplitude
T_power = (I^2) / (I_2 * I_3)

# Taylor expand the Power Overlap around dL = 0
approximation = T_power.subs({m:1, n:1}).taylor(dL, 0, 2)
show(approximation.simplify_full())
```
